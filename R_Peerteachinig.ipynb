## Arranging Plots

### visualize multiple datasets in one plot

```python
%%R
data1 <- data.frame(value = rnorm(100, mean = 0, sd = 1))
data2 <- data.frame(value = rnorm(100, mean = 1, sd = 1))


ggplot() +
  geom_histogram(data = data1, aes(x = value, y = ..density..),
                 binwidth = 0.5, fill = "skyblue", alpha = 0.7, position = "identity") +
  geom_histogram(data = data2, aes(x = value, y = ..density..),
                 binwidth = 0.5, fill = "coral", alpha = 0.7, position = "identity") +
  labs(title = "Überlagerte Histogramme", x = "Wert", y = "Dichte")
```
